```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# forward bias
fb_table = pd.DataFrame({
    "Voltage (V)": [0.5, 0.7, 0.8, 0.9, 1.0, 1.1, 1.2, 1.4, 1.5, 1.6, 2.0],
    "Current (mA)": [0, 0.4, 0.6, 1.0, 1.4, 1.8, 2.2, 2.6, 3.4, 3.9, 4.2]
})

print(fb_table)
```

```python
# reverse bias
rb_table = pd.DataFrame({
    "Voltage (V)": [0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 4.0, 5.0, 6.0],
    "Current (μA)": [0, 2, 2, 2, 2, 3, 4, 7, 11, 17]
})

print(rb_table)
```

## Visuals

```python
# forward bias
plt.figure()
plt.plot(fb_table['Voltage (V)'], fb_table['Current (mA)'], marker='o')
plt.xlabel('Forward Voltage (V)')
plt.ylabel('Forward Current (mA)')
plt.title('PN Diode Forward Bias Characteristics')
plt.grid(True)
plt.show()
```

```python
# reverse bias plot

plt.figure()
plt.plot(rb_table['Voltage (V)'], rb_table['Current (μA)'], marker='o')
plt.xlabel('Reverse Voltage (V)')
plt.ylabel('Reverse Current (μA)')
plt.title('PN Diode Reverse Bias Characteristics')
plt.grid(True)
plt.show()
```

```python
plt.figure() #unscaled 3rd quadrant

# forward bias - 1st quadrant
plt.plot(
    fb_table['Voltage (V)'],
    fb_table['Current (mA)'],
    marker='o',
    label='Forward Bias (mA)'
)

# rerverse bias - 3rd quadrant
plt.plot(
    -rb_table['Voltage (V)'],
    -rb_table['Current (μA)'],
    marker='s',
    label='Reverse Bias (μA)'
)

plt.axhline(0, linewidth=1)
plt.axvline(0, linewidth=1)

plt.xlabel('Voltage (V)')
plt.ylabel('Current')
plt.title('PN Diode V-I Characteristics')
plt.legend()
plt.grid(True)
plt.show()
```
